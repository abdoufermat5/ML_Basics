#### 3. Pour créer une image en mode RGB, il faut créer un vecteur de 3 dimensions dont la première dimension est la hauteur, la seconde est la largeur et la troisième représente les canaux RGB (chaque canal est représenté par une valeur de type numpy uint8).

<b>(a) Créer un vecteur numpy de taille (100, 300, 3) dont les trois canaux ont tous la valeur 0 pour tous les pixels (ce qui représente une image en noir),</b>

````python
W, H, Z = 100, 300, 3
black_image = np.zeros((W, H, Z))
plt.imshow(black_image)
````

````python
grey_img = np.zeros((W, H), dtype=np.uint8)
grey_img[:,:100] = 0
grey_img[:,100:200] = 150
grey_img[:,200:] = 255
plt.imshow(grey_img, cmap = 'Greys_r')
````

````python
img_ = Image.fromarray(np.uint8(grey_img))
img_
````
